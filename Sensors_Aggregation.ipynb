```python
sensor_bucket['start_date']=pd.to_datetime(sensor_bucket['start_date'],origin='unix').dt.floor('D')
sensor_bucket['end_date']=pd.to_datetime(sensor_bucket['end_date'],origin='unix').dt.floor('D')
sensor_bucket['latitude']=sensor_bucket['latitude'].round(2)
sensor_bucket['longitude']=sensor_bucket['longitude'].round(2)
sensor_bucket.head()
```

```python
keys=factors_filtered['ï»¿measurement_value']
values=factors_filtered['description']
sensors_dict = dict(zip(keys, values))

def sensor_names(val):
  if val in sensors_dict:
    return sensors_dict[val]
  else:
    return str(val)

print(sensors_dict)  
```

```python
sensor_bucket['measurement_value']=sensor_bucket['measurement_value'].apply(sensor_names)
sensor_bucket.head(20)
```
